```
from keras.applications.resnet50 import ResNet50, decode_predictions, preprocess_input
ResNetModel = ResNet50(weights='imagenet')
# ResNetModel.summary()

# Build a new model that is ResNet50 minus the very last layer
last_layer = ResNetModel.get_layer("avg_pool")
ResNet2048 = keras.Model(inputs=ResNetModel.inputs, outputs=last_layer.output)
# ResNet2048.summary()
```

```
def video2Matix(videoData, root):
  cap=cv2.VideoCapture(root)
  rawPrediction = []
  l = videoLength(root)
  for i in range(l):
  # while True:
    ret, frame=cap.read()
    if ret == False:
      break
    x = cv2.resize(frame, (224,224), interpolation = cv2.INTER_AREA)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    preds = ResNet2048.predict(x)
    rawPrediction.append(preds)

  rawPrediction = np.asarray(rawPrediction)

  fftData = []
  ResNetFFT(rawPrediction, fftData, fMax)

  videoData.append(fftData)
  return videoData
```

```
def videoLength(root):
  cap=cv2.VideoCapture(root)
  i = 0
  while True:
    ret, frame=cap.read()
    if ret == False:
      break
    i = i + 1
  return i
```

```
def ResNetFFT(rawPrediction, fftData, fMax):
  for attribute in range(2048):
    temp = []
    for i in range(len(rawPrediction)):
      temp.append(rawPrediction[i, 0, attribute])
    temp = np.asarray(temp)
    yf = fft(temp)
    OUT = np.abs(yf)
    fftData.append(OUT[:fMax])
  fftData = np.asarray(fftData)
  return fftData
```
